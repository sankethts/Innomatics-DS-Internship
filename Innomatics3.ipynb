```python
from typing import List
import heapq

class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        heap = [float('-inf'), float('-inf')]
        heapq.heapify(heap)

        for num in nums:
            if num > heap[0]:      # compare with smallest in heap
                heapq.heappop(heap)
                heapq.heappush(heap, num)

        return (heap[0] - 1) * (heap[1] - 1)

```

```python
sol = Solution()
print(sol.maxProduct([3, 4, 5, 2]))   
print(sol.maxProduct([1, 5, 4, 5]))     
```

```python
class Solution:
    def numTeams(self, rating: List[int]) -> int:
        ans, n = 0, len(rating)
        for i, b in enumerate(rating):
            l = sum(a < b for a in rating[:i])
            r = sum(c > b for c in rating[i + 1 :])
            ans += l * r
            ans += (i - l) * (n - i - 1 - r)
        return ans
```

```python
sol = Solution()

print(sol.numTeams([2, 5, 3, 4, 1]))   # Expected output: 3
print(sol.numTeams([2, 1, 3]))         # Expected output: 0
print(sol.numTeams([1, 2, 3, 4]))      # Expected output: 4
```

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
       
        count = 0
        for start, end in zip(startTime, endTime):
            if start <= queryTime <= end:
                count += 1
        return count
```

```python
sol = Solution()

startTime = [1, 2, 3]
endTime = [3, 2, 7]
queryTime = 4

print(sol.busyStudent(startTime, endTime, queryTime))  
```

```python
class Solution(object):
    def numberOfSteps (self, num):
        """
        :type num: int
        :rtype: int
        """
        result = 0
        while num:
            result += 2 if num%2 else 1
            num //= 2
        return max(result-1, 0)
```

```python
sol = Solution()

print(sol.numberOfSteps(14))
print(sol.numberOfSteps(8))   
print(sol.numberOfSteps(0))  
```

```python
class Solution:
    def countBits(self, n: int) -> List[int]:
        return [i.bit_count() for i in range(n + 1)]
```

```python
sol = Solution()

print(sol.countBits(2)) 
print(sol.countBits(5))
```
