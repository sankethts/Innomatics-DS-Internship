```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "marketing_data.csv"
df = pd.read_csv(file_path)
```

```python
df.head(), df.info()
```

```python
# Fix column name
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
# Basic preprocessing
df['Age'] = 2024 - df['Year_Birth']
```

```python
num_cols = ['Income', 'Age', 'Recency', 'MntWines']
cat_cols = ['Education', 'Marital_Status', 'Country']
```

```python
# Univariate Analysis (Numerical - Visual)
for col in num_cols:
    plt.figure()
    df[col].dropna().hist()
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Univariate Analysis (Categorical - Visual) 
for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Bivariate Analysis

# Numerical vs Numerical
plt.figure()
plt.scatter(df['Income'], df['MntWines'])
plt.xlabel('Income')
plt.ylabel('Wine Spending')
plt.title('Income vs Wine Spending')
plt.show()
```

```python
# Numerical vs Categorical
plt.figure()
df.boxplot(column='Income', by='Education')
plt.title('Income by Education')
plt.suptitle('')
plt.xlabel('Education')
plt.ylabel('Income')
plt.show()
```

```python
# Categorical vs Categorical
cross_tab = pd.crosstab(df['Marital_Status'], df['Response'])
cross_tab.plot(kind='bar')
plt.title('Marital Status vs Campaign Response')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```
